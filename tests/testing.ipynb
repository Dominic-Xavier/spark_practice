```python
def repeated(text, sub):
    words = text.replace('.', '').split()
    freq = {}

    for word in words:
        if word == sub:
            freq[sub] = freq.get(sub, 0) + 1
    return freq

str_x = "Emma is good developer. Emma is a writer"
substring = "Emma"

print(repeated(str_x, substring))
```
